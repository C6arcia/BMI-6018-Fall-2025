```python
#import pandas since data is set up as pandas dataframes
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class #class in AutoViz contains logic for detecting data types and generating plits automatically
```

```python
#AutoViz uses one dataframe for all features
#put x and y into one table 
df = pd.concat([X, y], axis=1)
df.head() #show first 5 rows 
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()

dft = AV.AutoViz(
    filename="",  #using dataframe directly
    dfte=df       #combined dataframe
)
```

```python
#plots did not show on the first run
#force to show plots plots inside the cell (Jupyter specific)
%matplotlib inline

from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class() #set up AutoViz to read dataset
dft = AV.AutoViz( #telling AutoViz data is contained in a file, rather a dataframe  
    filename="",
    dfte=df,
    depVar=""  #must be empty for full EDA
)
```
