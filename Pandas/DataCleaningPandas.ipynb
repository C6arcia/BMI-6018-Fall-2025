```python
flights_data = pd.read_csv(r'C:\Users\clara\Desktop\data\flights.csv')
flights_data.head(10)

#print(flights_data)
#Pandas Data Filtering/Sorting Question Answering
#use flights_data

#Question 1 How many flights were there from JFK to SLC? Int
q_1 = flights_data.loc[(flights_data['origin']=='JFK') & (flights_data['dest']=='SLC')]
print("q_1 =", len(q_1))

#Question 2 How many airlines fly to SLC? Should be int
q_2 = flights_data.loc[(flights_data['dest']=='SLC')]
print("q_2 =", len(q_2))

#Question 3 What is the average arrival delay for flights to RDU? float
RDU_flights = flights_data.loc[(flights_data['dest']=='RDU')]
RDU_avg_delay = RDU_flights['arr_delay'].mean() 
print("q_3 = average arrival delay for flights to RDU is", float(RDU_avg_delay))

#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
SEA_flights = flights_data.loc[(flights_data['dest']=='SEA')]
#print(len(SEA_flights))
#flights from LGA and JFK to SEA
#.isin() is a cleaner tool to use instead of writing seperate functions for LGA and JFK
NYC_to_SEA = SEA_flights[SEA_flights['origin'].isin(['LGA','JFK'])]
#print(len(NYC_to_SEA))
q_4 = len(NYC_to_SEA)/len(SEA_flights)
print("q_4 =", float(q_4))           

#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
#create a new date column in specified format
flights_data['date'] = flights_data.apply(lambda row: f"{int(row['year'])}/{int(row['month'])}/{int(row['day'])}", axis=1)
#group by date and calculate average departure delay
avg_dep_delay_by_date = flights_data.groupby('date')['dep_delay'].mean()
#find the date with the max avg departure delay
max_dep_delay_date = avg_dep_delay_by_date.idxmax()
max_dep_delay_value = avg_dep_delay_by_date.max()
#create a slice with date and float
q_5 = pd.DataFrame({'date': [max_dep_delay_date], 'avg_dep_delay': [float(max_dep_delay_value)]})
print("q_5 =", q_5)

#Question 6 Which date has the largest average arrival delay? pd slice with date and float
#group by date and calculate average arrival delay
avg_arr_delay_by_date = flights_data.groupby('date')['arr_delay'].mean()
#find the date with the max avg arrival delay
max_arr_delay_date = avg_arr_delay_by_date.idxmax()
max_arr_delay_value = avg_arr_delay_by_date.max()
#create a slice with date and float
q_6 = pd.DataFrame({'date': [max_arr_delay_date], 'avg_arr_delay': [float(max_arr_delay_value)]})
print("q_6 =", q_6)

#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
#filter flights from LGA or JFK in 2013
NYC_flights = flights_data[(flights_data['origin'].isin(['LGA', 'JFK'])) & (flights_data['year'] == 2013)]
#account for missing airtime values/ NA
NYC_flights = NYC_flights[NYC_flights['air_time'] > 0]
#calculate speed 
NYC_flights = NYC_flights.copy()  
NYC_flights['speed'] = NYC_flights['distance'] / NYC_flights['air_time']
#flight with the highest speed
fastest_flight = NYC_flights.loc[NYC_flights['speed'].idxmax()]
#create a slice with tailnumber and speed
q_7 = pd.DataFrame({'tailnum': [fastest_flight['tailnum']], 'speed': [float(fastest_flight['speed'])]})
print("q_7 =", q_7)
```

```python
weather_data_pd = pd.read_csv(r'C:\Users\clara\Desktop\data\weather.csv')
weather_data_np = weather_data_pd.to_numpy()

#Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
#replace nans with 0s
weather_data_pd = weather_data_pd.fillna(0)

#Numpy Data Filtering/Sorting Question Answering
#Use weather_data_np


#Question 9 How many observations were made in Feburary? Int
#convert weather data to np
weather_data_np = weather_data_pd.to_numpy()
#get the column for month
month_index = list(weather_data_pd.columns).index('month')
#filter rows where month is February or ==2
February_rows = weather_data_np[weather_data_np[:, month_index] == 2]
#count the number of observations in February 
February_count = len(February_rows)
print("q_9 = Number of observations made in February:", February_count)

#Question 10 What was the mean for humidity in February? Float
#get column for month and humid
month_index = list(weather_data_pd.columns).index('month')
humidity_index = list(weather_data_pd.columns).index('humid')
#filter rows where month is February or ==2
February_rows = weather_data_np[weather_data_np[:, month_index] == 2]
#humidity values for February
February_humidity = February_rows[:, humidity_index]
#mean humidity
mean_humidity_February = np.mean(February_humidity)
print("q_10 = Mean humidity in February:", float(mean_humidity_February))

#Question 11 What was the std for humidity in February? Float
#std for humidity in February
std_humidity_February = np.std(February_humidity)
print("q_11 = std humidity in February:", float(std_humidity_February))
```
